```python
c_cases = []
dates_t = []
```

```python
for date in dates:
    t_cases = 0
    date_case = df.loc[df['Date'] == date]
    for cases in date_case['Confirmed']:
        t_cases = t_cases + cases
    c_cases.append(t_cases)
    dates_t.append(date)
```

```python
covid_india_org = pd.DataFrame(list(zip(dates_t, c_cases)), 
               columns =['Date', 'Confirmed cases']) 
```

```python
x_train = []
y_train = []
for i in range(10,202):
    x_train.append(covid_india_org['Confirmed cases'][i-10:i]) 
    y_train.append(covid_india_org['Confirmed cases'][i])
x_train, y_train = np.array(x_train), np.array(y_train) 
```

```python
x_train
```

```python
len(y_train)
```

```python
x_train.shape
```

```python
x_train = np.reshape(x_train,(x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor = Sequential()
```

```python
regressor.add(LSTM(units = 20, return_sequences = True, input_shape = (x_train.shape[1], 1) ))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 20, return_sequences = True ))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 20, return_sequences = True ))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 20))
regressor.add(Dropout(0.2))
```

```python
regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressor.fit(x_train, y_train, epochs = 10 , batch_size = 16)
```
